```python
import pandas as pd
import numpy as np
from sklearn.decomposition import LatentDirichletAllocation
import pyLDAvis
from pyLDAvis.sklearn import prepare

pyLDAvis.enable_notebook()
```

```python
from specs import restore_state
from topic_modeling import doc_term_frequency_matrix
```

```python
specs = restore_state()
```

```python
for path, spec in specs.items():
    new_freqs = dict()
    for code, freq in spec.code_frequencies.items():
        if code.startswith('128b') or code.startswith('256b') or code.startswith('512b'):
            new_freqs[code] = freq
    specs[path].code_frequencies = new_freqs
```

```python
docs, terms, mat = doc_term_frequency_matrix(specs.values())
```

```python
lda = LatentDirichletAllocation(n_components=10)
components = lda.fit_transform(mat)
```

```python
class TermsListAsVectorizer:
    '''
        Is used to make a list of terms look and feel like
        sklearn.feature_extraction.text.(CountVectorizer, TfIdfVectorizer)
        for the `pyLDAvis`.
    '''

    def __init__(self, arr):
        self.arr = arr

    def get_feature_names(self):
        return self.arr
```

```python
prepare(lda, np.asmatrix(mat), TermsListAsVectorizer(terms))
```
